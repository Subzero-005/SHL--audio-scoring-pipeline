```python
import pandas as pd
import numpy as np
import os

print("Searching for uploaded submission files...")
# Automatically find the files wherever Kaggle put them
v2_path = None
v3_path = None

for dirname, _, filenames in os.walk('/kaggle'):
    for filename in filenames:
        if filename == 'submission_v2.csv':
            v2_path = os.path.join(dirname, filename)
        elif filename == 'submission_v3 (1).csv':
            v3_path = os.path.join(dirname, filename)

if not v2_path or not v3_path:
    print("ERROR: Could not find one or both CSV files. Please make sure they are uploaded and named exactly 'submission_v2.csv' and 'submission_v3 (1).csv'")
else:
    print(f"Found V2 NLP: {v2_path}")
    print(f"Found V3 DL: {v3_path}")
    
    # Read the files
    v2_nlp = pd.read_csv(v2_path) 
    v3_dl = pd.read_csv(v3_path) 

    print("Blending Predictions (70% Deep Learning / 30% NLP)...")
    # Calculate the weighted average
    blended_predictions = (v3_dl['label'] * 0.70) + (v2_nlp['label'] * 0.30)

    # Build the final submission file
    submission_blend = pd.DataFrame({
        'filename': v3_dl['filename'],
        'label': np.clip(blended_predictions, 0, 5)
    })

    submission_blend.to_csv('submission_final_blend.csv', index=False)
    print("\nSUCCESS! submission_final_blend.csv is ready. Check your output panel!")
    display(submission_blend.head())
```
